## Задание 1: Зеркальное отражение цифр

Пользователь вводит пятизначное число. Программа должна зеркально отразить три центральные цифры. Первая и последняя остаются на местах.

<br>

**Примеры:**
* `23456 -> 25436`
* `30789 -> 38709`

```python
str = input("Enter 5 digits")
if (len(str)!=5):
    print("wrong length !=5")
    exit(-1)

strL = list(str)
tmp = strL[1]
strL[1] = strL[3]
strL[3] = tmp
print(''.join(strL))
```

## Задание 2: Подсчет выходных до отпуска

Пользователь вводит сколько дней осталось до ближайшего отпуска. Программа должна вывести количество выходных дней до отпуска, если учесть, что выходные это суббота и воскресенье, сегодня понедельник и праздники мы не учитываем.  

<br>

**Примеры:**
* `4  -> 0`
* `6  -> 1`
* `14 -> 4`

```python
try:
    count = int(input())
except:
    print("Only digits supported")
    exit(-1)
weeks = count // 7
days = count % 7
dayoff = weeks * 2 + ((7 -days) if days > 5 else 0)
print(dayoff)  
```

## Задание 3: Плитка шоколада

Пользователь вводит длину и ширину плитки шоколада, а также размер куска, который хочет отломить. Программа должна вычислить - можно ли совершить подобный разлом или нет, если учесть, что ломать можно только по прямой.

<br>

**Примеры:**
* `3, 4, 6  -> True`
* `5, 7, 8  -> False`
* `4, 5, 12 -> True`

```python
str = input("Enter 3 digits splitted by comma:")
l = str.split(",")
if (len(l)!=3):
    print("Enter 3 digits")
    exit(-1)
try:
    w = int(l[0].strip())
    h = int(l[1].strip())
    p = int(l[2].strip())
except ValueError:
    print("only digits supported")
    exit(-1)

s = w * h
if (s%p) ==0 :
    print(True)
else:
    print(False)
```

## Задание 4: Римские числа

Пользователь вводит целое положительное число, программа должна вернуть строку в виде римского числа.

<br>

**Примеры:**
* `3   -> III`
* `15  -> XV`
* `234 -> CCXXXIV`

```python
def RomeDigits(count : int, div1):
    if div1 !=100 and div1 != 10 and div1 != 1:
        raise ValueError 
    match count:
        case 1:
            match div1:
                case 100:
                    return "C"
                case 10:
                    return "X"
                case 1:
                    return "I"
        case 2:
            match div1:
                case 100:
                    return "CC"
                case 10:
                    return "XX"
                case 1:
                    return "II"
        case 3:
            match div1:
                case 100:
                    return "CCC"
                case 10:
                    return "XXX"
                case 1:
                    return "III"
        case 4:
            match div1:
                case 100:
                    return "CD"
                case 10:
                    return "XL"
                case 1:
                    return "IV"
        case 5:
            match div1:
                case 100:
                    return "D"
                case 10:
                    return "L"
                case 1:
                    return "V"
        case 6:
            match div1:
                case 100:
                    return "DC"
                case 10:
                    return "LX"
                case 1:
                    return "VI"
        case 7:
            match div1:
                case 100:
                    return "DCC"
                case 10:
                    return "LXX"
                case 1:
                    return "VII"
        case 8:
            match div1:
                case 100:
                    return "DCCC"
                case 10:
                    return "LXXX"
                case 1:
                    return "VIII"
        case 9:
            match div1:
                case 100:
                    return "CM"
                case 10:
                    return "XC"
                case 1:
                    return "IX"
    return ""
try:
    num = int(input())
except ValueError:
    print("only digits supported")
    exit(-1)
if (num > 1000):
    print("numbers more than 1000 not supported")
    exit(-1)

hundreds = num // 100
dozens = (num - hundreds*100) // 10
eds = num -hundreds*100 - dozens *10
try:
    print(RomeDigits(hundreds,100) + RomeDigits(dozens, 10) + RomeDigits(eds,1))
except ValueError:
    print("wrong diveder")
    
```

## Задание 5: Проверка на вещественное число

Пользователь вводит данные, проверить - являются ли они положительным вещественным числом. Не использовать встроенные функции для проверки, только методы данных и конструкцию IF. (Дополнительное задание, по желанию - проверка на отрицательные вещественные числа)

<br>

**Примеры:**
* `5.6  -> True`
* `.78  -> True`
* `.67. -> False`
* `5    -> True`

```python
import re

str1 = input("Enter float number")
valid = re.compile(r"\d*\.\d*")

print(str1, " is float  - ", True if valid.fullmatch(str1) else False)
```
