## Задание 1: Конвертер регистров

Написать функцию, которая будет переводить snake_case в PascalCase и наоборот. 

Функция должна сама определять - какой формат ей передали. Можно добавить ключевой аргумент, который будет принудительно возвращать один из форматов.

<br>

**Примеры:**
* `otus_course     -> OtusCourse`
* `PythonIsTheBest -> python_is_the_best`



```python
def ChangeCase(in_str: str ):
    result = ''
    idx = 0
    pos = in_str.find("_")
    camelCaseExist = True
    if pos != -1:
        camelCaseExist = False
    while idx < len(in_str):
        ch =str(in_str[idx])
        if ch.isupper():
            if not camelCaseExist:
                print("snake_case expected but there is ipper case symbol")
                raise ValueError
            else:
                ch = ("_" + ch.lower()) if idx != 0 else ch.lower()
        elif ch == "_":
            idx += 1
            if idx >= len(in_str):
                break
            ch = in_str[idx]
            if not camelCaseExist:
                ch = ch.upper()
            else:
                print("camelCase expected but there is \"_\" symbol")
                raise ValueError
        else:
            if not camelCaseExist and idx == 0:
                ch = ch.upper()
        result += ch
        idx +=1
    return result

in_str = input()

print(ChangeCase(in_str))
        

```

* `31.04.1962 -> False`



```python
import datetime
import re

def ParseDateString(in_str: str):
    pattern = r"(\d{2}).(\d{2}).(\d{4})"
    m = re.search(pattern, in_str)
    if not m:
        print("Wrong date format dd.mm.yyyy expected")
        raise ValueError
    day = int(m.group(1))
    month = int(m.group(2))
    year = int(m.group(3))
  
    try:
        dt = datetime.datetime(year,month,day)
        return True
    except ValueError:
        print("Wrong date entered.")
        return False


in_str = input()

print(ParseDateString(in_str))
    
```

## Задание 3: Проверка на простое число

Функция проверки на простое число. Простые числа – это такие числа, которые делятся на себя и на единицу.

<br>

**Примеры:**
* `17 -> True`
* `20 -> False`
* `23 -> True`

```python
import math

def IsSimple(number: int)->bool:
    if number <=1:
        return False
    if number < 4:
        return True
    
    if (number %2 ==0 or number % 3 == 0):
        return False
    
    idx = 5
    max_diveder = int(math.sqrt(number)) + 1 
    while idx < max_diveder:
        if (number % idx == 0 or number % (idx +2) == 0 ):
            return False
        idx += 6
    return True

number = int(input())
print(IsSimple(number))
```
